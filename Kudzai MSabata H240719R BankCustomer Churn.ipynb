```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import BaggingClassifier
from sklearn.ensemble import AdaBoostClassifier
from sklearn.metrics import classification_report, roc_auc_score
```

```python
BankCustomer=pd.read_csv(r"Bank Customer Churn Prediction.csv")
```

```python
BankCustomer.head()
```

```python
BankCustomer.isnull().sum()
```

```python
BankCustomer.dtypes
```

```python
BankCustomer["gender"].replace({"Female":1,"Male":0},inplace=True)
```

```python
BankCustomer.head()
```

```python
for col in ["country"]:
    BankCustomer[col]=LabelEncoder().fit_transform(BankCustomer[col])
```

```python
BankCustomer.head()
```

```python
BankCustomer2=BankCustomer.drop(columns=["customer_id"],axis=1)
```

```python
BankCustomer2.head()
```

```python
BankCustomer2.dtypes
```

```python
x=BankCustomer2.drop(["churn"],axis=1)
y=BankCustomer2["churn"]
```

```python
x_test,x_train,y_test,y_train=train_test_split(x,y,test_size=0.2)
```

```python
LR_Model=LogisticRegression()
LR_Model.fit(x_train,y_train)
RF_Model=RandomForestClassifier()
RF_Model.fit(x_train,y_train)
AB_Model=AdaBoostClassifier()
AB_Model.fit(x_train,y_train)
BC_Model=BaggingClassifier()
BC_Model.fit(x_train,y_train)
DTC_Model=DecisionTreeClassifier()
DTC_Model.fit(x_train,y_train)
```

```python
pred=LR_Model.predict(x_test)
pred1=RF_Model.predict(x_test)
pred2=AB_Model.predict(x_test)
pred3=BC_Model.predict(x_test)
pred4=DTC_Model.predict(x_test)
```

```python
print(classification_report(y_test,pred))
print(classification_report(y_test,pred1))
print(classification_report(y_test,pred2))
print(classification_report(y_test,pred3))
print(classification_report(y_test,pred4))
```

```python
#Conclusion: Random Forest Classifier and Adaboost perfomed well  with accuaracy level of 85% .AdaBoost is slightly more aggressive with a better recall. Random forest is slightly more conservaive(better precision).Therefore Adaboost is the best  as it improves recall and F1 Score of the minority class
```
